# Usage of PiecePoly Module

This notebook demonstrates the usage of the `PiecePoly` module, which constructs a piecewise polytropic equation of state (EoS) for a neutron star. 

The module allows us to calculate pressure and energy density over different segments of a neutron star based on polytropic relations.

```julia
# 1. Define the crust parameters and polytropic indices
log_p1 = 34.380
Gamma = [2.227, 2.189, 2.159]  # Example polytropic indices

# 2. Calculate the parameters for the equation of state
rho_lim_all, a_all, K_all, Gamma_all = PiecePoly.get_all_params(log_p1, Gamma)
println("Density Limits: ", rho_lim_all)
println("a parameters: ", a_all)
println("K parameters: ", K_all)
println("Gamma parameters: ", Gamma_all)

# 3. Create the piecewise polytropic equation of state
, p = PiecePoly.make_polyEos(rho_lim_all, a_all, K_all, Gamma_all)
# Now you can further process the eos data, plot, or use it for further analysis
plot(.*gcm3_to_MeVfm3, p.*gcm3_to_MeVfm3;
    xscale=:log10,
    yscale=:log10,
    label="Example Eos",
    legend=:topleft,
)
```
